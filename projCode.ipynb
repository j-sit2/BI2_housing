```python
#Examine data: Use pandas + read.csv to load data and create dataframe
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
path = 'projdata/canhouse-spending.csv'
df = pd.read_csv(path)
df.head()
```

```python
#Clean dataset: first check for missing values and what under column they fall under
df.isnull().sum()
```

There is missing data in the tobacco categories due to how before 2019, the category was slightly more narrow in scope, only factoring tobacco and alcohol. This category would eventually be succeeded by the tobacco, alcohol, and non-medicinal cannabis category. To streamline this data preprocessing stage, the results for these 2 separate categories will be merged into a single column.

```python
#Clean dataset: replace missing data with median from same data
df['Tobacco products, alcoholic beverages (and cannabis for non-medical use in 2019)'] = df['Tobacco products and alcoholic beverages (Terminated)']
df['Tobacco products, alcoholic beverages (and cannabis for non-medical use in 2019)'].fillna(1785, inplace=True)
df = df.drop(['Tobacco products and alcoholic beverages (Terminated)', 'Tobacco products, alcoholic beverages and cannabis for non-medical use'], axis=1)
df.head(10)
```

```python
path2 = 'projdata/canhouse-industryincome.csv'
df2 = pd.read_csv(path2)
df2.head(21)
```

```python
df2.isnull().sum()
```

There is missing data in a few categories here. However, the missing data is found in years that are outside the scope of the comparison outlined for this project. Therefore, these rows can be dropped.

```python
df2 = df2.drop(labels=range(0,9), axis=0)
df2 = df2.drop(labels=[19,20], axis=0)
```

```python
df2.isnull().sum()
```

```python
df2.head(10)
```

```python
df2.min()
```

```python
df2["Mining, quarrying, and oil and gas extraction"] = pd.to_numeric(df2["Mining, quarrying, and oil and gas extraction"])
df2["Utilities"] = pd.to_numeric(df2["Utilities"])
df2["Information and cultural industries"] = pd.to_numeric(df2["Information and cultural industries"])
df2["Finance and insurance"] = pd.to_numeric(df2["Finance and insurance"])
df2["Professional, scientific and technical services"] = pd.to_numeric(df2["Professional, scientific and technical services"])
df2["Management of companies and enterprises"] = pd.to_numeric(df2["Management of companies and enterprises"])
df2["Public administration"] = pd.to_numeric(df2["Public administration"])
```

```python
df2['produce3xfoodval'] = df2['Accommodation and food services']*3
```

```python
df2['produce3xfood'] = df2['Goods producing industries'] >= df2['produce3xfoodval']
df2['produce3xfood'].head(10)
```

```python
df2['miningsum3val'] = df2['Accommodation and food services'] + df2['Arts, entertainment and recreation'] + df2['Administrative and support, waste management and remediation services']
```

```python
df2['miningsum3'] = df2['Mining, quarrying, and oil and gas extraction'] >= df2['Accommodation and food services'] + df2['Arts, entertainment and recreation'] + df2['Administrative and support, waste management and remediation services']
df2['miningsum3'].head(10)
```

```python
df2['mostsimilar2val'] = df2['Finance and insurance'] - df2['Information and cultural industries']
```

```python
df2['mostsimilar2'] = df2['Finance and insurance'] >= df2['Information and cultural industries']
df2['mostsimilar2'].head(10)
```
